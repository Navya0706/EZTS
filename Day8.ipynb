```python
class BSTNode:
    def __init__(self, data):
        self.value = data
        self.left = None
        self.right = None

preo = []
posto = []
ino = []

def preorder(root):
    if root is None:
        return
    preo.append(root.value)
    preorder(root.left)
    preorder(root.right)

def inorder(root):
    if root is None:
        return
    inorder(root.left)
    ino.append(root.value)
    inorder(root.right)

def postorder(root):
    if root is None:
        return
    postorder(root.left)
    postorder(root.right)
    posto.append(root.value)

def insert(root, key):
    if root is None:
        return BSTNode(key)
    else:
        if root.value < key:
            root.right = insert(root.right, key)
        else:
            root.left = insert(root.left, key)
    return root

if __name__ == "__main__":
    root = None
    keys = [4, 2, 7, 1, 3, 6, 9]

    for key in keys:
        root = insert(root, key)

    preorder(root)
    inorder(root)
    postorder(root)

print("preorder-", preo)
print("inorder-", ino)
print("postorder-", posto)
```
